```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import sys
import time
import copy
import string
import json
import pickle as pk
import math

from treelib import Node, Tree

from collections import Counter

from math import sin, asin, cos, radians, fabs, sqrt

############################    从文件中 读取 数据   ##############
def read_q(filename):
    f = open(filename,"rb+")
    st = pk.load(f)
    f.close()
    return st

##########################    向文件中  写 数据  #####################
def write_q(st, filename):
    f = open(filename, "wb+")
    pk.dump(st,f)
    f.close()
```

```python
###########################  比较 path 相似性  #########################
# 5个文件中所有结点 路径相似性的平均值

big_t = [

[	"s_cluster"	,	"0_1"	,	"r_2"	,	3	,	2	],
[	"s_cluster"	,	"0_1"	,	"r_3"	,	3	,	3	],
[	"s_cluster"	,	"0_1"	,	"r_4"	,	3	,	4	],
[	"s_cluster"	,	"0_1"	,	"r_5"	,	3	,	5	],
[	"s_cluster"	,	"0_1"	,	"r_6"	,	3	,	6	],
[	"s_cluster"	,	"0_2"	,	"r_2"	,	3	,	2	],
[	"s_cluster"	,	"0_2"	,	"r_3"	,	3	,	3	],
[	"s_cluster"	,	"0_2"	,	"r_4"	,	3	,	4	],
[	"s_cluster"	,	"0_2"	,	"r_5"	,	3	,	5	],
[	"s_cluster"	,	"0_2"	,	"r_6"	,	3	,	6	],
[	"s_cluster"	,	"0_3"	,	"r_2"	,	3	,	2	],
[	"s_cluster"	,	"0_3"	,	"r_3"	,	3	,	3	],
[	"s_cluster"	,	"0_3"	,	"r_4"	,	3	,	4	],
[	"s_cluster"	,	"0_3"	,	"r_5"	,	3	,	5	],
[	"s_cluster"	,	"0_3"	,	"r_6"	,	3	,	6	]

]

pro_l = []
i = 0

for bb in big_t:
    cluster = bb[0]   # 聚类器  a_cluster,,mb_kmeans
    ex_times = bb[1]      #第几次实验
    at = bb[2]
    #print(cluster, ex_times, at)

    wz = [47,79,137,146,167]

    pro = 0
    num = 0

    for aa in wz:
        j = aa
    #print("当前处理文件为: ",j)    
        filename1 = ".\\os_file\\res_8\\"+str(cluster)+"\\"+str(ex_times)+"\\"+str(at)+ "\\tra_paths_"+str(i)+"_"+str(j)+".pkl"   
        filename2 = ".\\os_file\\res_8\\"+str(cluster)+"\\"+str(ex_times)+"\\"+str(at)+ "\\tra_rec_paths_"+str(i)+"_"+str(j)+".pkl"
        t_l = read_q(filename1)
        r_l = read_q(filename2)
        #print("当前文件原始轨迹：",t_l)
        #print("当前文件重构轨迹：",r_l)
    
        #print("当前文件轨迹长度：",len(r_l))
        num += len(t_l)
        for k in range(len(t_l)):   # 对于文件内的每一个叶子结点，对应的路径list
            #print("t_l[k]: k=: ",k,t_l[k])
            #print("r_l[k]: k=: ",k,r_l[k])
            len_max = max(len(t_l[k]), len(r_l[k]))
            len_min = min(len(t_l[k]), len(r_l[k]))
        
            si_len = 0
            for l in range(len_min):
            
                if t_l[k][l] == r_l[k][l]:
                    continue
                else:
                    si_len = l 
                    break
                
            if si_len == 0:
                si_len = len_min
            
        #print("si_len: ",si_len)            
            si_pro = round(si_len/len_max, 5)
        #print("si_pro",si_pro)
        
            pro += si_pro
        
    pro = round(pro/num,5)
    pro_l.append(pro)
    
pro_l
```
