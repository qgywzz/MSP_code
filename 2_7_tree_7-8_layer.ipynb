```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import sys
import time
import copy
import string
import json
import pickle as pk

from treelib import Node, Tree
from math import sin, asin, cos, radians, fabs, sqrt

############################    从文件中 读取 数据   ##############
def read_q(filename):
    f = open(filename,"rb+")
    st = pk.load(f)
    f.close()
    return st

##########################    向文件中  写 数据  #####################
def write_q(st, filename):
    f = open(filename, "wb+")
    pk.dump(st,f)
    f.close()
```

```python
filename = ".\\os_file\\st_poi_clusters\\s_cluster\\st_poi_cluster_3nd.pkl"
st_poi = read_q(filename)
```

```python

#####################   7 层树  ---把 dict-list 结构 转化为 tree   #####################

filename = ".\\os_file\\st_poi_clusters\\s_cluster\\semantic_tree_7.pkl"

a = 0
for item1 in st_poi:
    
    print("第一大类： ", item1)
    for item2 in st_poi[item1]:
        b = 0
        b = a
        st = st_poi[item1][item2]    # st_poi["enter"]["bar"]    st 表示第四层
        print("     第二大类： ", item2,"该条目下共有： ",len(st), " 个list")
        
#############    第一层    ###################################################
        if len(st) == 0:
            continue
        
        elif len(st)== 4 and type(st[0]).__name__ == 'list':        # 如果长度为 4 且下一级为list，则 遍历 建中间结点
            
            print(item2, "为第一次K-means建立中间结点, “i”长度为： ",len(st))
            for i in range(len(st)):        #####   *************   第一层    
                ######   st[j]
                
                name1 = item1+ "_" + item2 + "_" + str(i)
                st_pt.create_node(name1,name1, parent=item2)
                
                st_1 = st[i]       # 下一层
                
################  第二层  ################# 判断的是K2， 第二次K-means   ##############################
                if len(st_1) == 0:
                    continue
        
                elif len(st_1)== 4 and type(st_1[0]).__name__ == 'list':
                                                
                    print(item2, "为第二次K-means建立中间结点, “j”长度为： ",len(st_1))
                    for j in range(len(st_1)):      ########   ********    第二层
                        
                        name2 = name1 +"_"+str(j)
                        st_pt.create_node(name2,name2, parent = name1)
                        
                        st_2 = st_1[j]
                        
###############  第三层  #################################################################
                        if len(st_2) == 0:
                            continue
        
                        elif len(st_2)== 4 and type(st_2[0]).__name__ == 'list':
                                                
                            print(item2, "为第三次K-means建立中间结点, “k”长度为： ",len(st_2))
                            for k in range(len(st_2)):      ########   ********    第二层
                        
                                name3 = name2 +"_"+str(k)
                                st_pt.create_node(name3,name3, parent = name2)
                        
                                st_3 = st_2[k]
                        
###################   第四层   #########################################################################
                        ###############   最后一层  #####################
                                if len(st_3) == 0:
                                    continue
                            
                                else:
                                    print(item2, "第三次K-means后，建立叶子结点, “n”长度为： ",len(st_3))
                                    a += len(st_3)
                                    for n in range(len(st_3)):
                                        name4 = name3+"_"+str(n)
                                        st_pt.create_node(name4,name4,data = st_3[n], parent = name3)
                   
                    ########################################################

##################   第四层   #########################################################################
                    
                        else:       # 下一级为 字典，建叶子结点    第二层
                            print(item2, "第二次K-means后，建立叶子结点, “k”长度为： ",len(st_2))
                            a += len(st_2)
                            for k in range(len(st_2)):
                                name3 = name2 +"_"+str(k)
                                st_pt.create_node(name3,name3,data = st_2[k], parent = name2)
                    
####################   第三层   ######################################################################   
                
                else:       # 下一级为 字典，建叶子结点    第二层
                    print(item2, "第一次K-means后，建立叶子结点, “j”长度为： ",len(st_1))
                    a += len(st_1)
                    for j in range(len(st_1)):
                        name2 = name1 +"_"+str(j)
                        st_pt.create_node(name2,name2,data = st_1[j], parent = name1)
                
####################   第二层   #####################################################################
            
        else:       # 下一级为 字典，建叶子结点     第一层
            print(item2, "为item2，直接建立叶子结点，“i”长度为： ",len(st))
            a += len(st)
            for i in range(len(st)):
                name1 = item1+ "_" + item2 + "_" +str(i)
                st_pt.create_node(name1,name1,data = st[i], parent=item2)
                
            #tree = con_leaf(st, name2)
#####################    第一层  ###################################################################       
            
        print(item2,"共建立叶子结点： ",a-b," 个")
print("共建立叶子结点： ",a,"个")
            
write_q(st_pt, filename)
    
```
